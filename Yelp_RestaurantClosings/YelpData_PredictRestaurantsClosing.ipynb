## Discard non-Food & Beverage Businesses
I limited the scope of this project to **restaurants, bars, and cafes**. 
- I want to get rid of non-restaurants/cafes/bars, so I discarded rows that didn't contain food-related categories. I chose the categories in the `categories_to_keep` list (below), because they were the food-related `categories` within the top 50 most prevalent categories. 
- Discard businesses that are linked to food/beverage categories, but whose primary function is *NOT* to serve as a restaurant/bar/cafe (e.g., Walmart): 
    - Rows that are categorized as **'Drugstore'** and **'Gas Stations'** (because those tended to be Walmarts, Targets, etc, and gas stations)
    - Rows that are only categorized as **'Grocery;Food'** and **'Food;Grocery'** (and no other category)

```python
yelp_food = yelp_og # use copy of original (just in case)

# Food/Beverage categories: 
# (Note: These categories were selected based on top 50 most prevelant categories)
categories_to_keep = ['Food', 'Restaurant', 'Bars', 'Dessert', 'Coffee', 'Cafe', 'Breakfast']
categories_to_discard = ['Drugstore', 'Gas Stations']

# 1. Keep only rows that contain any of the categories specified in categories_to_keep list:
category_checker_keep = lambda row_categories: any(word in row_categories for word in categories_to_keep)
yelp_food = yelp_food[yelp_food.categories.apply(category_checker_keep) == True]

# 2. Discard all rows that contain any of the categories specified in the categories_to_discard list:
category_checker_discard = lambda row_categories: any(word in row_categories for word in categories_to_discard)
yelp_food = yelp_food[yelp_food.categories.apply(category_checker_discard) == False]

# 3. Get rid of anything that is only categorized as a Grocery store and nothing else
yelp_food = yelp_food[~((yelp_food['categories']=='Grocery;Food') | (yelp_food['categories']=='Food;Grocery'))]
    
print(yelp_food.shape)
```

### Treatment of Outliers for `latitude` and `longitude`:

**Outliers that are erroneous:** There appear to be errors in latitude and longitude data. I looked at the range of lat/long data within cities, and found that it was well outside of what would be expected. The latitude/longitude data is sometimes inconsistent with the other location data (e.g., address) given for businesses (e.g., max latitude = 90 - that's basically the north pole for a bar that was supposed to be in England). 
- **Confirming that `latitude` and `longitude` variables where the error lies:** Cross corroboration of random sample of outliers on google maps confirms that these errors are most likely in the latitude and longitude data.
- **Solution:** *20 degrees outside city+state median*: Replace with city+state median latitude/longitude (this will also take care of other potential inconsistencies/errors).  
    - Note: 20 degrees away is often in another US state (or country if we're talking about Europe), and thus seemed like a reasonable cut-off.  

**Geographical Outliers:** Most of the data points are concentrated in North America and Europe. However, there is a subset of data points (*n* = 23) that are in the southern hemisphere (i.e., South America). 
**Create broad geographical regional categorical variable instead of using longitude and latitude**
- Based on histogram; solution to non-normality of data;

```python
### Set latitude/longitude outliers (20 degrees above/below city+state median) to NaN:
# Note: Using the df_train to derive transformation parameters for all variables

# Create city/state medians:
city_latlong_sum = df_train[['city', 
                              'state', 
                              'latitude', 
                              'longitude']].groupby(['city','state'], 
                                                    as_index=False).agg({'latitude': ['median', 'std', 'count', 'min', 'max'],
                                                                         'longitude': ['median', 'std', 'count', 'min', 'max']})
# Flatten column names for merging:
city_latlong_sum.columns = ['_'.join(col) for col in city_latlong_sum.columns]
city_latlong_sum = city_latlong_sum.rename({"city_": "city", "state_": "state"}, 
                                           axis='columns')

# Merge city/state summary with **FULL** dataset
latlong_recode = pd.merge(df_full[['business_id', 'city', 'state', 'latitude', 'longitude']], 
                          city_latlong_sum[['city', 'state', 
                                            'latitude_median', 'longitude_median']],
                           how = 'left')
print(latlong_recode.shape)

# Set outlying (+-20 degrees above/below city+state median) lat/long values to 
# the median lat/long for that city+state:
latlong_recode['latitude'] = latlong_recode['latitude'].mask(latlong_recode['latitude'] < (latlong_recode['latitude_median']-20), 
                                                            latlong_recode['latitude_median'])
latlong_recode['latitude'] = latlong_recode['latitude'].mask(latlong_recode['latitude'] > (latlong_recode['latitude_median']+20), 
                                                            latlong_recode['latitude_median'])
latlong_recode['longitude'] = latlong_recode['longitude'].mask(latlong_recode['longitude'] < (latlong_recode['longitude_median']-20), 
                                                            latlong_recode['longitude_median'])
latlong_recode['longitude'] = latlong_recode['longitude'].mask(latlong_recode['longitude'] > (latlong_recode['longitude_median']+20), 
                                                            latlong_recode['longitude_median'])

# Isolate training rows within temporary training set:
latlong_rec_tr = latlong_recode.iloc[0:56260,]
print(latlong_rec_tr.shape)

# Verify that it worked: (check out our friend at the North Pole)
latlong_recode[latlong_recode.business_id == '1yQUqh3_h1IOrXZmb4CBFw']
```

### Latitude and longitude Bands:
- Split into **Quartiles** ordinal categories: 
    - *Con:* Fails to distinguish how far away the South American locations are
    - *Pro:* Does a better job chopping up the regions where most of the sample is. 

```python
### Latitude: Categorize based on quartiles
pd.qcut(latlong_rec_tr['latitude'], 4).value_counts().sort_index(ascending=False)
```

```python
### Longitude: Categorize based on quartiles
pd.qcut(latlong_rec_tr['longitude'], 4).value_counts().sort_index()
```

```python
### Recode latitude and longitude as quartiles using functions:

def bin_latitude(df): 
    #df.latitude = df.latitude.fillna()
    bins = (-89.99, 35.274, 41.311, 43.708, 90) # -90 and 90 = possible range of latitude
    group_names = [0, 1, 2, 3]
    categories = pd.cut(df.latitude, bins, labels=group_names)
    df['latQ'] = categories
    df['latQ'] = df['latQ'].astype(int)
    return df

def bin_longitude(df): 
    bins = (-179.99, -111.931, -80.832, -79.388, 180)
    group_names = [0, 1, 2, 3]
    categories = pd.cut(df.longitude, bins, labels=group_names)
    df['longQ'] = categories
    df['longQ'] = df['longQ'].astype(int)
    return df

# Meta-function:
def bin_latlong(df):
    df = bin_latitude(df)
    df = bin_longitude(df)
    return df

latlong_recode = bin_latlong(latlong_recode)
latlong_recode.head()
```

### Code 'chain_size' and 'review_count' variables:
- Make chain variable binary: 0 = not a chain, 1 = chain
- Bin review_count into 6 bins (using pd.qcut)

```python
## Bin review_count based on df_train
pd.qcut(df_train.review_count, 6).value_counts().sort_index()
```

```python
### Bin review_count variable
def bin_review_count(df): 
    bins = (0, 4, 8, 15, 30, 73, float('inf')) 
    group_names = [0, 1, 2, 3, 4, 5]
    categories = pd.cut(df.review_count, bins, labels=group_names)
    df['rev_count_bin'] = categories
    df['rev_count_bin'] = df['rev_count_bin'].astype(int)
    return df

### Make binary version of chain size variable to help control outliers
def dummy_chain(df): 
    df['chain_dummy'] = df['chain_size'].map(lambda s: 1 if s > 1 else 0)
    return df

# Meta-function:
def recode_countdata(df):
    df = bin_review_count(df)
    df = dummy_chain(df)
    return df

df_full = recode_countdata(df_full)

df_train = update_df_train(df_full) # Update training dataset

print(df_full.shape)
print(df_train.shape)
```

```python
def plot_distribution( df , var , target , **kwargs ):
    row = kwargs.get( 'row' , None )
    col = kwargs.get( 'col' , None )
    facet = sns.FacetGrid( df , hue=target , aspect=4 , row = row , col = col )
    facet.map( sns.kdeplot , var , shade= True )
    facet.set( xlim=(df[ var ].min(), df[ var ].max() ) )
    facet.add_legend()

def plot_correlation_map( df , varList):
    corr = df_train[varList].corr()
    _ , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    _ = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
    )
```
